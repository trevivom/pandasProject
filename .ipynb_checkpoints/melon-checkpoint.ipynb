```python
url = 'http://www.melon.com/chart/index.htm'
driver.get(url)
```

```python
from bs4 import BeautifulSoup
# 전체 html 코드 갖고 오기 (html)
html = driver.page_source
# 사람이 보기 좋게 정리하기. (soup)
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

```python
songs = soup.select('tr')[1:]
print(len(songs))
print(songs[0])
```

```python
song = songs[0]
```

```python
title = song.select('a')
len(title)
```

```python
title = song.select('div.ellipsis.rank01 > span > a')
title
```

```python
title = song.select('div.ellipsis.rank01 > span > a')[0].text
title
```

```python
singer = song.select('div.ellipsis.rank02 > a')[0].text
singer
```

```python
# 노래 100개 제목, 가수 출력하기
for song in songs:
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select('div.ellipsis.rank02 > a')[0].text
    print(title, singer, sep = '|')
```

```python
song_data = []
rank = 1

songs = soup.select('tr')[1:]
for song in songs:
    title = song.select('div.ellipsis.rank01 > span > a')[0].text
    singer = song.select('div.ellipsis.rank02 > a')[0].text
    song_data.append(['Melon', rank, title, singer])
    # append 는 이어붙인다. 라고 생각하자.
    rank = rank + 1

song_data
```

```python
# 데이터분석 라이브러리 pandas 가져오기
import pandas as pd
# 엑셀 파일에서 칼럼 이름 정리
column = ['서비스', '순위', '타이틀', '가수']
# 데이터 프레임 생성하기 // columns 는 데이터 속성이라 생각하자.
pd_data = pd.DataFrame(song_data, columns = column)
# 생성된 데이터 프레임 확인하기
pd_data.head()
```

```python
# 엑셀 파일 추출하기
pd_data.to_excel('./files/melon.xlsx', index=False)
```
